```python
# Smart way of checking if the syntax works on column

# It works

pd.to_datetime(df.head(5)['Date'])
```

```python
# Changing Date column into datetime64[ns]

df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Create WEEKDAY column

df['Weekday'] = df['Date'].map(lambda x : x.isoweekday())
```

```python
#Create TIME column

df['Hour'] = df['Date'].map(lambda x: x.hour)
```

```python
#Create MONTH column

df['Month'] = df['Date'].map(lambda x: x.month)
```

```python
#Change YEAR column into date time
df['Year'] = df['Date'].map(lambda x: x.year)
```
